```python
import pandas as pd
import numpy as np
import glob
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
def load_tennis_data_file(filepath):
    if filepath.endswith('.csv'):
        df = pd.read_csv(filepath)
    else:
        df = pd.read_excel(filepath)
    
    df = df.rename(columns={
        'Tournament': 'tourney_name',
        'Date':       'tourney_date',
        'Surface':    'surface',
        'Round':      'round',
        'Winner':     'winner_name',
        'Loser':      'loser_name',
        'WRank':      'winner_rank',
        'LRank':      'loser_rank',
        'Best of':    'best_of'
    })
    
    keep = ['tourney_name','tourney_date','surface','round',
            'winner_name','loser_name','winner_rank','loser_rank','best_of']
    df = df[[c for c in keep if c in df.columns]]
    df['tourney_date'] = pd.to_datetime(df['tourney_date'], errors='coerce')
    df['source'] = 'tennis_data'
    return df

data_folder = r'C:\Users\paulc\OneDrive\Projects\wimbledon-predictor\data'

excel_files = glob.glob(os.path.join(data_folder, '*.xlsx')) + \
              glob.glob(os.path.join(data_folder, '*.xls')) + \
              glob.glob(os.path.join(data_folder, '2025*.csv'))

print(f"Files found: {len(excel_files)}")
for f in sorted(excel_files):
    print(os.path.basename(f))

td_dfs = [load_tennis_data_file(f) for f in sorted(excel_files)]
td_combined = pd.concat(td_dfs, ignore_index=True)
print(f"\nTennis Data files loaded: {len(td_dfs)} files, {len(td_combined)} rows")
td_combined.head()
```

```python
df_2024 = pd.read_csv(os.path.join(data_folder, 'atp_matches_2024.csv'))

df_2024['tourney_date'] = pd.to_datetime(df_2024['tourney_date'].astype(str), 
                                          format='%Y%m%d', errors='coerce')
df_2024['best_of'] = df_2024.get('best_of', 3)

keep = ['tourney_name','tourney_date','surface','round',
        'winner_name','loser_name','winner_rank','loser_rank','best_of']
df_2024 = df_2024[[c for c in keep if c in df_2024.columns]]
df_2024['source'] = 'sackmann'

print(f"2024 loaded: {len(df_2024)} rows")
df_2024.head()
```

```python
all_matches = pd.concat([td_combined, df_2024], ignore_index=True)

for col in ['surface', 'winner_name', 'loser_name', 'tourney_name', 'round']:
    if col in all_matches.columns:
        all_matches[col] = all_matches[col].astype(str).str.strip().str.lower()

all_matches = all_matches.dropna(subset=[
    'tourney_date', 'surface', 'winner_name', 
    'loser_name', 'winner_rank', 'loser_rank'
])

all_matches['year'] = all_matches['tourney_date'].dt.year
all_matches = all_matches.sort_values('tourney_date').reset_index(drop=True)

print(f"Total matches: {len(all_matches)}")
print(f"Years: {all_matches['year'].min()} - {all_matches['year'].max()}")
print(f"Surfaces: {all_matches['surface'].unique()}")
```

```python
grass = all_matches[all_matches['surface'] == 'grass'].copy().reset_index(drop=True)
print(f"Grass court matches: {len(grass)}")
print(f"Wimbledon matches: {grass[grass['tourney_name'].str.contains('wimbledon', na=False)].shape[0]}")

save_path = os.path.join(data_folder, 'grass_combined_2000_2025.csv')
grass.to_csv(save_path, index=False)
print(f"Saved to: {save_path}")
grass.head()
```
